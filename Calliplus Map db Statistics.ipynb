# Calliplus db manipulation

# 載入資料庫 

```python
import pandas as pd
import sqlite3

con=sqlite3.connect("map.db")

cur=con.cursor()

dw = pd.read_sql_query("""SELECT chardatas.character,authors.author,charbooks.charbook, chardatas.img_url 
                                            from chardatas, authors,charbooks 
                                            where chardatas.author = authors.id 
                                                        and chardatas.charbook=charbooks.id""",
                       con)

df = pd.DataFrame(dw)

img_prefix = "http://www.cns11643.gov.tw/FontManager/FontManager?id="

df.url = img_prefix + df.IMG_URL

df['url'] = img_prefix + df['IMG_URL']
```
